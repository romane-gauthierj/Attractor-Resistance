```python
import pandas as pd 
from io import StringIO
```

```python
mutations_data = pd.read_csv('data/cellmodel_data/mutations_all_20250318.csv')
```

```python
# genes_diff_expressed_data = pd.read_csv('analysis/Refametinib/genes_models/MEK1_MEK2_target_block/results/genes_diff_expressed/significant_genes_EGF_ON_INVASION.csv')

genes_lines = []
with open('analysis/Refametinib/genes_models/MEK1_MEK2_target_block/results/genes_diff_expressed/significant_genes_EGF_ON_INVASION.csv', "r") as file:
    for line in file:
        if line.strip().startswith('--- Summary'):
            break  # Stop reading when you hit the summary line
        genes_lines.append(line)
```

```python
genes_clean = [row.strip() for row in genes_lines]
data_str = "\n".join(genes_clean)

genes_diff_expressed = pd.read_csv(StringIO(data_str))
```

```python

with open("analysis/Refametinib/top_resistant_ids.txt", "r") as file:
    patients_res_ids = [line.strip() for line in file]

with open("analysis/Refametinib/top_sensitive_ids.txt", "r") as file:
    patients_sens_ids = [line.strip() for line in file]

with open("analysis/Refametinib/top_healthy_ids.txt", "r") as file:
    patients_healthy_ids = [line.strip() for line in file]
```

```python
intersection = set(patients_healthy_ids) & set(patients_sens_ids) & set(patients_res_ids)
print(intersection)
```

```python
patients_ids =  patients_healthy_ids + patients_sens_ids + patients_res_ids
patients_ids
print(len(patients_ids))
```

```python
genes_diff_expressed.head()
```

```python
genes_diff_expressed = list(set(genes_diff_expressed['Unnamed: 0']))
```

```python
mutations_data_filtered = mutations_data[
    (mutations_data['model_id'].isin(patients_ids)) & 
    (mutations_data['gene_symbol'].isin(genes_diff_expressed))
]

mutations_data_filtered
```

```python
mutations_res = mutations_data_filtered[
    (mutations_data['model_id'].isin(patients_res_ids)) 
]

mutations_sens = mutations_data_filtered[
    (mutations_data['model_id'].isin(patients_sens_ids)) 
]

mutations_healthy = mutations_data_filtered[
    (mutations_data['model_id'].isin(patients_healthy_ids)) 
]
```
